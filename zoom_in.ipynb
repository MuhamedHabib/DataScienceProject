# <h2><span style= 'color:noire'>1- preprocessing : </span></h2>
---

```python
import cv2 
import matplotlib.pyplot as plt

```

```python
image_path='1.jpg'
image = cv2.imread(image_path)
```

```python
# convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(image)
```

```python
plt.imshow(gray)
```

```python
# create a binary thresholded image
_, binary = cv2.threshold(gray, 130, 255, cv2.THRESH_BINARY_INV)
# show it
plt.imshow(binary, cmap="gray")
plt.show()
```

```python
# find the contours from the thresholded image
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# draw all contours
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
plt.imshow(image)
```

# <h2><span style= 'color:noire'>2- determiner coordonnees contour : </span></h2>
---

```python
x,y=[],[]
for box in contours:
    x.append(box[0][0][0])
    y.append(box[0][0][1])
```

```python
import numpy as np 
```

```python
x = np.array(x)
y= np.array(y)
```

```python
values, counts = np.unique(y, return_counts=True)
```

```python
l_val_y=values.tolist()
l_count_y=counts.tolist()
```

# <h2><span style= 'color:noire'>4- le contour a x0 y0 x1 y1 les plus frequent dans les valeur : </span></h2>
---

```python
def list_max_val(lst_count,lst_val):
    
    it=0
    max_val_count=[]
    index_max=[]
    for i in lst_count:
        if i==max(lst_count):
            index_max.append(it)
        it=it+1
    for i in index_max:
        max_val_count.append(lst_val[i])
        
    return max_val_count
```

```python
val_x=list_max_val(l_count_x,l_val_x)
val_y=list_max_val(l_count_y,l_val_y)  
```

```python
val_x
```

```python
val_y
```

# <h2><span style= 'color:noire'>5- largeur et hauteur contour si plus 4 point grace image mal traiter  : </span></h2>
---

```python
largeur=max(val_x)-min(val_x)
hauteur=max(val_y)-min(val_y)
```

# <h2><span style= 'color:noire'>6- facteur de multiplication est shape image / min(largeur,hauteur) : </span></h2>
---

```python
img = cv2.imread(image_path)
```

```python
if largeur!=0 and hauteur!=0:
    coef_resize_x=int(img.shape[1]/largeur)
    coef_resize_y=int(img.shape[0]/hauteur)
    factor=max(coef_resize_x,coef_resize_y)
elif largeur==0:
    factor=img.shape[0]/hauteur
elif hauteur == 0:
    factor=img.shape[1]/largeur
    
factor
```

# <h2><span style= 'color:noire'>7- resize de l image pour zoomer : </span></h2>
---

```python
def zoom(img, zoom_factorx,zoom_factory):
    return cv2.resize(img, None, fx=zoom_factorx, fy=zoom_factory)
```

```python
zoomed = zoom(img,factor,factor)
```

```python
cv2.imshow('z',zoomed)
cv2.waitKey(0)
cv2.imwrite('z.jpg',zoomed)
```
